```python
#f(x) =0.6x^3+5.5x^2+10x-5
#Определить корни
#1. Найти интервалы, на которых функция возрастает
#2. Найти интервалы, на которых функция убывает
#3. Построить график
#4. Вычислить вершину
#5. Определить промежутки, на котором f > 0
#6.Определить промежутки, на котором f < 0

from sympy import *
a = 0.6
b = 5.5
c = 10
d = -5

x = Symbol('x')

f_x = a*(x**3) + b*(x**2) + c*(x) + d
```

```python
f_x
```

```python
d=diff(f_x)
d
```

```python
from sympy import plot

graph = plot(d,f_x,(x,-10,5), legend = True, Xlable = 'X', Ylable='Y', title='График функции и её производной')

```

```python
print("корень №1:", solve(f_x,x)[0].n(3))
print("корень №2:", solve(f_x,x)[1].n(3))
print("корень №3:", solve(f_x,x)[2].n(3))
```

```python
print("Абсцисса вершины №1:", solve(d,x)[0].n(3))
print("Абсцисса вершины №2:", solve(d,x)[1].n(3))
```

```python
print(f'Функция больше 0 на промежутках: {solve(f_x > 0,x)}')
```

```python
print(f'Функция меньше 0 на промежутках: {solve(f_x < 0,x)}')
```

```python
print(f'Функция возрастает на промежутках: {solve(d > 0,x)}')
```

```python
print(f'Функция убывает на промежутках: {solve(d < 0,x)}')
```
